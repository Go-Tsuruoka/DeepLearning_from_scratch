# 活性化関数
今までは，入力と重みの積の総和にバイアスを加えたものが0以上なら1，0未満なら0を出力するような関数を採用していた．つまり，以下のようになっていた．
$$y = \left\{ \begin{array}{ll} 1 & (b + \sum_{i=1}^n w_ix_i) \geq 0 \\ 0 & (b + \sum_{i=1}^n w_ix_i) < 0 \end{array} \right.$$

ここに以下のような$h(x)$を採用しよう．
$$h(x) = \left\{ \begin{array}{ll} 1 & x \geq 0 \\ 0 & x < 0 \end{array} \right.$$

この$h(x)$を**活性化関数**と呼ぶ．すると，今までのパーセプトロンは以下のように書ける．
$$y = h(b + \sum_{i=1}^n w_ix_i)$$

先ほどの$h(x)$は，ステップ関数と呼ばれる．実際のニューラルネットワークでは，ステップ関数ではなく，ステップ関数の微分可能な関数や，sigmoid関数など様々な関数が用いられる．

以下ではこれらの活性化関数とその入出力関係をグラフにしてみる．

## ステップ関数
今までのニューラルネットワークで用いた，以下の関数$h(x)$をステップ関数と呼ぶ．
$$h(x) = \left\{ \begin{array}{ll} 1 & x \geq 0 \\ 0 & x < 0 \end{array} \right.$$

実際にグラフにしてみる．

```python
import numpy as np
import matplotlib.pylab as plt
# ステップ関数

def step_function(x):
    return np.array(x>0,dtype=int)
x=np.arange(-5.0,5.0,0.1)
y=step_function(x)
plt.plot(x,y)
plt.ylim(-0.1,1.1)
plt.xlabel("input")
plt.ylabel("output")
plt.show()
```

## シグモイド関数
次にシグモイド関数を見てみる．シグモイド関数は以下のように定義される．
$$h(x) = \frac{1}{1 + e^{-x}}$$

```python
import numpy as np
import matplotlib.pylab as plt
# ステップ関数

def sigmoid_function(x):
    return 1/(1+np.exp(-x))

x=np.arange(-5.0,5.0,0.1)
y=sigmoid_function(x)
plt.plot(x,y)
plt.ylim(-0.1,1.1)
plt.xlabel("input")
plt.ylabel("output")
plt.show()
```

## ReLU関数
近年使われる活性化関数の一つとしてReLU関数がある．ReLU関数は以下のように定義される．
$$h(x) = \left\{ \begin{array}{ll} x & x \geq 0 \\ 0 & x < 0 \end{array} \right.$$

```python
import numpy as np
import matplotlib.pylab as plt
# ステップ関数

def ReLU(x):
    return np.maximum(0,x)
x=np.arange(-5.0,5.0,0.1)
y=ReLU(x)
plt.plot(x,y)
plt.ylim(-0.1,5.1)
plt.xlabel("input")
plt.ylabel("output")
plt.show()
```
